### 개인정보 비식별화

```python
import re

def mask_personal_info(text):

    #특정 규칙에 의거하여 회사명 마스킹
    company_regex = r'.*(은행|증권|카드)'
    text = re.sub(company_regex, r'**\1', text)

    #개인/고객센터 전화번호 마스킹
    phone_regex = r'\b\d{2,3}-\d{3,4}-\d{4}\b'
    text = re.sub(phone_regex, 'XXX-XXXX-XXXX', text)

    cust_phone_regex = r'\b\d{4}-\d{4}\b'
    text = re.sub(cust_phone_regex, 'XXXX-XXXX', text)

    #띄어쓰기 없는 회사 번호 마스킹
    call_regex = r'\b\d{10}\b'
    text = re.sub(call_regex, 'XXX-XXX-XXXX', text)

    #특정 URL 마스킹
    url_regex = r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\\(\\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'
    text = re.sub(url_regex, '[URL]', text)
    
    #규칙 기반 고객명 마스킹
    name_regex = r"([가-힣][^\s님]*)\s?(고객)?님"
    text = re.sub(name_regex, r'*** 고객님', text)

    #KB 특정 어플/서비스 마스킹
    # kb_regex = r'(KB)\s?(스타뱅킹|Pay|금융쿠폰|스타클럽|마이데이터|Wallet|금융|PAY)'
    kb_regex = r"(KB|신한|삼성|하나)[ \t]*[a-zA-Z가-힣]*[ \t]*[a-zA-Z가-힣]*[ \t]*"
    text = re.sub(kb_regex, r'[Service]', text)
    
    predefined_info = ['하나원큐', '한국투자', '네이버페이', '신한', '삼성']
    for word in predefined_info:
        text = text.replace(word,'[Company]')
    return text

deidentified_df = []

for text in df_all:
    deidentified_text = mask_personal_info(text)
    deidentified_df.append(deidentified_text)
```

### 비식별화된 텍스트 -> 한 줄 개요 생성 (GPT4 API 사용)

```python
import openai
from tqdm import tqdm
from nltk import sent_tokenize
oneline_template = """
[마케팅 메시지 시작]
{예시}
[마케팅 메시지 끝]

당신은 마케팅 메시지를 분석하고 해당 메시지의 1줄 개요를 작성해주는 AI입니다.
위 메시지는 날짜, 경품 등이 구체적으로 명시되어있습니다. 저는 이러한 구체적인 내용들을 모두 비식별화한 후 일반화시킨 개요를 얻고 싶습니다.
위에 제공된 마케팅 메시지를 읽은 후 해당 메시지의 1줄 개요를 작성해주세요.


아래 나열된 지침들을 기반으로 생성해주세요.
1. 전체적인 마케팅 메시지의 핵심 내용(이벤트 방식)을 중심으로 작성해라.  
2. 위에 제공된 마케팅 메시지에 작성되지 않은 내용에 대해 작성하지 않는다.
"""

sentence_template = """
[마케팅 메시지 시작]
{예시}
[마케팅 메시지 끝]

당신은 한국어에 능통한 언어학자 AI입니다.
위에 제공된 마케팅 메시지를 꼼꼼하게 읽은 후 해당 메시지를 완전한 문장 형태로 작성해주세요.

아래 나열된 지침을 기반으로 생성해주세요.
1. 위에 제공된 마케팅 메시지에 작성되지 않은 내용에 대해 작성하지 않는다.
"""


VALID_CHATGPT_MODELS = ['gpt-3.5-turbo', 'gpt-4']


def get_instructions():
    instructions_dict = {
        'oneline' : oneline_template,
        'sentence' : sentence_template
    }
    return instructions_dict

def set_openai_key(key):
    openai.api_key = key

class AskQuestions():

    def __init__(self, text, mode, model='gpt-3.5-turbo'):
        self.text = text
        self.model = model
        self.mode = mode
        self.instruction = get_instructions()

    def ask_question(self):
        instruction = self.instruction[self.mode]
        chatgpt_messages = prepare_chatgpt_message(
            instruction,
            self.text
        )
        response = call_chatgpt(chatgpt_messages, model=self.model)

        return response

def prepare_chatgpt_message(task_prompt, text):
    input_prompt = task_prompt.replace('{예시}', text)
    messages = [{"role": "system", "content": input_prompt}]
    return messages

def call_chatgpt(chatgpt_messages, model="gpt-3.5-turbo"):
    response = openai.ChatCompletion.create(model=model, messages=chatgpt_messages, temperature=0, max_tokens=1024)
    reply = response['choices'][0]['message']['content']
    return reply

open_ai_key = 'chatgpt api key'
set_openai_key(open_ai_key)
```

```python
# 비식별화된 텍스트를 기반으로 gpt4 API를 활용하여 개요 생성
def auto_oneline_generate(df):
    response_list = []
    model = 'gpt-4'
    for i in tqdm(range(len(df))):
        text = df[i]
        chat_try = AskQuestions(text= text, model=model, mode='oneline')
        response = chat_try.ask_question()
        response_list.append(response)
    return response_list
oneline_list = auto_oneline_generate(deidentified_df)
```

```python
# 비식별화된 텍스트를 기반으로 gpt3.5 API를 활용하여 문장 단위로 쪼개기
def auto_sentence_generate(df):
    response_list = []
    model = 'gpt-3.5-turbo'
    for i in tqdm(range(len(df))):
        text = df[i]
        chat_try = AskQuestions(text= text, model=model, mode='sentence')
        response = chat_try.ask_question()
        for sent in sent_tokenize(response):
            response_list.append(sent)
    return response_list
sentences_list = auto_sentence_generate(deidentified_df)
```

```python
# 비식별화된 텍스트들 
all_sentences = pd.DataFrame(sentences_list)
all_sentences.to_csv('Dataset/kb_dataset_all_sentences.csv',index_label=False)
```

```python
# 비식별화된 텍스트와 gpt4로 생성한 한줄 개요를 함께 dataframe형태로 저장
kb_dataset_deidentified_overview = pd.DataFrame(zip(deidentified_df, oneline_list), columns=['deidentified_text', 'overview'])
kb_dataset_deidentified_overview.to_csv('Dataset/kb_dataset_deidentified_oneline_overview2.csv',index_label=False)
```
